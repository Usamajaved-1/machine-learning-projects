```python
data.isnull().sum()
# data["fare"].fillna(data["fare"].mean(),inplace = True)
```

```python
# ab hm data ki nature dakhain ga 
sns.distplot(data["fare"])
plt.show()
```

```python
data.describe() # outlier present ha ya nhi confirm kr lain ga mazeed
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(data[["fare"]]) # special focus [[]] these brackets
```

```python
ss.transform(data[["fare"]])
```

```python
data["fare_ss"] = pd.DataFrame(ss.transform(data[["fare"]]),columns= ["x"])
# fare_ss ak new column aga sara opr wala hi ha last ma direction di ha x axis ma 
data.head(3)
```

```python
# ab isko dakhain ga  kya std 1 ha or data kam howa b ha ya nhi mean magnitude/size kam howa ha ?
data.describe()
```

```python
# ab isko grphical representation sa dakhain ga 
plt.figure(figsize= (12,4))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(data["fare"])

plt.subplot(1,2,2)
plt.title("aftter")
sns.distplot(data["fare_ss"])

plt.show()
```

# min max/normalization scalling 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
data = sns.load_dataset("titanic")
data.head(3)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(data[["fare"]])
mms.transform(data[["fare"]])
```

```python
data["fare_mms"] =pd.DataFrame(mms.transform(data[["fare"]]),columns= ["x"])
data.head(3) # same wasa hi jasa sb opr kiya ha 
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title("old")
sns.distplot(data["fare"])

plt.subplot(1,2,2)
plt.title("new")
sns.distplot(data["fare_mms"])
plt.show()
```
