```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv("placement-dataset.csv")
data.head(3)
```

```python
data.drop(columns="Unnamed: 0",inplace= True)
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(x="cgpa",y ="iq",hue= "placement",data= data)
plt.show()
```

```python
# it is linearlly seprable 
x = data.iloc[:,:-1]
y = data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.2,random_state=21)
```

```python
from sklearn.svm import SVC # SUPPort vector classifier bcz data is linear 
```

```python
sv = SVC(kernel="linear") # kernal is linear bcz line is best for our data explain in register 
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
# if trainnning is greater then overfit if testing is greater underfit 
```

```python


plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv("updated_placement.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(x="cgpa",y ="package",data= data)
plt.show()
```

```python
# it is linearlly seprable 
x = data[["cgpa"]]
y = data[["package"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.2,random_state=21)
```

```python
from sklearn.svm import SVR # SUPPort vector classifier bcz data is linear 
```

```python
sv = SVR(kernel="linear")
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
# now we want to seee the pred line 
plt.figure(figsize=[4,3])
sns.scatterplot(x="cgpa",y ="package",data= data)
plt.plot(data["cgpa"],sv.predict(x),data=data,color = "red")
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv("HR.csv")
data.head(3)
```

```python
data.drop(columns="Position",inplace=True)
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
data.columns = data.columns.str.strip() 
# It looks like the column name for "Salary" has leading and trailing spaces, which is why you're getting the error. You can fix this by stripping the spaces from the column names. Here's how you can do it:
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(x="Level",y ="Salary",data= data)
plt.show()
```

```python
# it is linearlly seprable 
x = data[["Level"]]
y = data[["Salary"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.2,random_state=21)
```

```python
from sklearn.svm import SVR # SUPPort vector classifier bcz data is linear 
```

```python
sv = SVR(kernel="sigmoid") # kernal is linear bcz line is best for our data explain in register 
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(x="Level",y ="Salary",data= data)
plt.plot(data["Level"],sv.predict(x),color = "red")
plt.show()
```
