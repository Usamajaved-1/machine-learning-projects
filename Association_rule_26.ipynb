```python
market = []
for i in range (0,data.shape[0]): # it selects all the rows from dataset 
    customer = []
    for j in data.columns: # it selects all the columns from dataset 
        if type(data[j][i])==str:  # if type of j and i are str them append it into custumer 
            customer.append(data[j][i])
    market.append(customer)
```

```python
# now there are two list in market list innside list but for counter we have te give only one lst 
list = []
for i in market:
    # print(i)
    for j in i :
        print(j) # by this only one list is created 
        list.append(j)      

# market = [["apple", "banana"], ["carrot", "date"], ["eggplant"]] this helps to understand better
# after this it become list = ["apple", "banana", "carrot", "date", "eggplant"]

```

```python
p = collections.Counter(list)
p
```

```python
p.keys(),p.values() # now thesea re seprate from each other but we have to create a data frame 
```

```python
# now we will create a dictinary 
d = {"items": p.keys(),"values":p.values()}
```

```python
# now we create a data frame 
df = pd.DataFrame(d).sort_values(by=["values"],ascending=False) 
df
# sort values by values is its argument by which it sorted the values according to our values column 
# assending false mean we want the max 
```
