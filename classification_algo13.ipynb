# Logistic_regression for Single Input 

```python
# single input logisti regression 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv("networkadds.csv")
data.head(3)
```

```python
# due to binomial we only take 2 columns age and purchases  
data.drop(columns=["User ID","Gender","EstimatedSalary"],inplace=True) # drop column which we dont need 
data.head(3)
```

```python
# now check it is applicable for logistic regression or not 
plt.figure(figsize=[5,3])
sns.scatterplot(x="Age",y="Purchased",data=data)
plt.show() 
```

```python
# our data follows the logistic regression 
# now sperate the independent and dpendent variables 
x = data[["Age"]] 
y = data["Purchased"]
x.ndim,y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[60]]) # no 
```

```python
# ab hm dakhain ga ka line kasa plot ho g
plt.figure(figsize=[5,3])
sns.scatterplot(x="Age",y="Purchased",data=data)
sns.lineplot(x="Age",y=lr.predict(x),data=data,c = "red")
plt.show() 
```

```python
# this line is explain in register and we are succesfull 
```

# Linear Regression for Multiple Inputs 

```python
# single input logisti regression 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv("placement-dataset.csv")
data.head(3)
```

```python
data.drop(columns=["Unnamed: 0"],inplace=True)
data.head(3)
```

```python
#  now check it linearlly seprable or not 
plt.figure(figsize=[5,3])
sns.scatterplot(x="cgpa", y = "iq",data = data)
plt.show()
```

```python
# now how to seprate these x and y so we use hue 
#  now check it linearlly seprable or not 
plt.figure(figsize=[5,3])
sns.scatterplot(x="cgpa", y = "iq",data = data,hue="placement")
plt.legend(loc = 1)
plt.show() # now our graph is linearlly seprable 
```

```python
x = data.iloc[:,:-1]
y = data["placement"]
x.ndim,y.ndim
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,train_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_train.shape,y_train.shape # 2 and 1 are columns 20 and 20 are rows 
```

```python
lr.score(x_test,y_test)*100  # not better but good according to testing point of view 
```

```python
lr.predict([[6.8,123.0]]) # approximately equals to one 
```

```python
plt.figure(figsize=[5,3])
sns.scatterplot(x="cgpa", y = "iq",data = data,hue="placement")
sns.lineplot(x="cgpa", y = lr.predict(x),data = data,c = "red")
plt.legend(loc = 1)
plt.show()
```

```python
# this method doesnot work in binomial so we have to take decsion boundry we have to import another boudry
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = lr)
plt.show()
# it require x y (both must be array like shape mean it become numpy array) and
# clf (classifier object / which model we use)
```

```python
# now you can see how the line intercept 
```

```python
# our equation become m1x1+m2x2+b  two inputs thats why x1and x2 and b is constant 
```

```python
lr.coef_,lr.intercept_
```

# classify through polynomial feature 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv("polynomial_binary_output_dataset.csv")
data.head(3)
```

```python
sns.scatterplot(x="data1",y="data2",data= data,hue = "output")
plt.legend(loc = 1)
plt.show()
```

```python
x = data.iloc[:,:-1]
y = data["output"]
x.shape
```

```python
# now we use polynomial fetaure
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
```

```python
pf.fit(x)
```

```python
x = pd.DataFrame(pf.transform(x))
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression() # max iterations 
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100 # see the accuracy is increase so much as compared to old one 
```
