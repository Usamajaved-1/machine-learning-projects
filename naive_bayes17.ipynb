```python
# naive bayes algo applied on the conditional probability for ex we train the model that if rain happen i dont do to watch movie or if not then i wil go it works on classification and algo both 
# naive : it represents features we give in the dataset  eplain on register
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv("placement-dataset.csv")
data.head(3)
```

```python
data.drop(columns="Unnamed: 0",inplace=True)
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
# now we see that how it is seprable linear or other 
sns.scatterplot(x = "cgpa",y = "iq",hue = "placement",data=data)
plt.show()
```

```python
# now see it is normal or discrete distribution 
plt.figure(figsize=[3,3])
sns.kdeplot(x = "cgpa", data=data)
plt.show() 
```

```python
# now see it is normal or discrete distribution 
plt.figure(figsize=[3,3])
sns.kdeplot(x = "iq", data=data)
plt.show() 
```

```python
x = data.iloc[:,:-1]
y = data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB # explain inregister
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100 ,gnb.score(x_train,y_train) *100
```

```python
ml = MultinomialNB()
ml.fit(x_train,y_train)
```

```python
ml.score(x_test,y_test)*100,ml.score(x_train,y_train)*100
```

```python
# poor accuracy is because we know it is for text prediction explain on register 
```

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
# poor bcz it is for bollean 
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plt.figure(figsize=[6,3])
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb) # all explain in previous files 
plt.show()
```

```python
# see how much accurate it is 
```

```python
plt.figure(figsize=[6,3])
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb) # all explain in previous files 
plt.show()
```

```python
plt.figure(figsize=[6,3])
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=ml) # all explain in previous files 
plt.show()
```
