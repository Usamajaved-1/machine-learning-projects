```python
import pandas as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.DataFrame({"Size":["s","m","m","l","s","xl","s","l"]})
data.head(3)
# ab hm encoding krain ga uska liye hma ak numbers decide krna para ga ka hm ksi order ma krna chah raha hain
order = [["s","m","l","xl"]]
from sklearn.preprocessing import OrdinalEncoder #yahn import kr liya 
# oe = OrdinalEncoder() # JO B IMPORT kiya usko call krwana to must ha na 
# lakin iski cattagery by default alpahbatical order ma hoti ha mean a to z wise kra ga so hma change kra ga uProcessLookupError
oe = OrdinalEncoder(categories= order) # ye hmna apna hisab sa adjust kr liya 
oe.fit(data[["Size"]]) # fit usko dakha ga or samjha ga or aklearn ka algo ko apply kr ka 
oe.transform(data[["Size"]]) #YAHN USki encoding ho g casesensitive
```

```python
data["Size_en"] = oe.transform(data[["Size"]]) # ak new column bna diya encoding ka or uska sath likh diya 
data
```

```python
# ye hmna sklearn ka through encoding kr li h 
# ab hm dakhain ga map function ka through encoding kasa kr sakta ha 
ord_data1 = {"s":0,"m":1,"l":2,"xl":3}
data["Size"].map(ord_data1) # data ma size ka andr map kr diya oder data 1 ka according
data["en_data_map"] = data["Size"].map(ord_data1)
data # yw hmari encoding ho gay 
```

# ab hm krain ga bara data set pa ordinal encoding

```python
import pandas as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
data.head(3)
```

```python
from sklearn.preprocessing import OrdinalEncoder
order_1 = [["Thur","Fri","Sat","Sun"]]
oe = OrdinalEncoder(categories= order_1)
oe.fit(data[["day"]])
```

```python
oe.transform(data[["day"]])
```

```python
data["day_en"] = oe.transform(data[["day"]])
data
```
