```python
# ab hm dakhain ga ka mean absolute division direct kasa find kr sakta ha
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sec_a=np.array([75,65,73,68,72,76])
sec_b=np.array([90,47,43,96,93,51])
num = np.array([1,2,3,4,5,6])
mean_a = sec_a.mean()
mean_b = sec_b.mean()
# mad isma hoti ha num - mean
b = sec_a-mean_a
c = sec_b-mean_b
# ab iska absolute valuae
np.abs(b)
np.abs(c)
# ab hm krain ga inka sum or phr divide kr da ga num of valuaes pa
mad_a = np.sum(np.abs(b))/len(sec_a)
mad_b = np.sum(np.abs(c))/len(sec_b)
mad_a,mad_b
# standard deviation and variance jiska kam ho ga hm us data ko prefer krain ga 
np.std(sec_a),np.std(sec_b)
np.var(sec_a),np.var(sec_b)
```

```python
# corelation example
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
# print(data)
data.isnull().sum()
# ab hmm covarience nikala ga uska liye hma data type ka pta hona chaheye 
# data.info()
# is sa hma data type pta chal jay g 
cor = data.select_dtypes(["float64","int64"]).corr()
# ab agr hmna isko show rain ga through heatmap
# sns.heatmap(cor)
# agr ab hmna ispa valuae b print karwani ho t 
sns.heatmap(cor,annot=True)
plt.show()
# ye hmna totalbill or size ka la liya 
```

```python
# corelation 0 to 1 ma lie kr ra ha
# hm asa hi co variance b nikl sakta ha 
cov = data.select_dtypes(["float64","int64"]).cov()
sns.heatmap(cov,annot=True)
plt.show()
# jbka iski value boht zyada ha 
```
