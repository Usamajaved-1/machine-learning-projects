# DBSCAN

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from  sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=250,noise=0.5) # store values from data set 
```

```python
x,y
```

```python
# now we create a data fraame x is our input have 2 col and y is our output 
```

```python
# we create a dictionary 
# df = {"data1":x[:,0],"data2":x[:,1],"output":y}
df = {"data1":x[:,0],"data2":x[:,1]} # we remove output due to 0 1 issue in dbscan predic below
# x[:,0] all rows of 1 col
# x[:,1] all rows of 2 col
```

```python
# now we create a data frame and store it in dataset 
data = pd.DataFrame(df)
data.head(3)
```

```python
# now checkit is linear or non linear 
import seaborn as sns
sns.scatterplot(x= "data1",y="data2",data=data)
plt.show()
```

```python
# so this is non linear dataste  
# NOW WE USE dbscan 
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.4,min_samples=5) # eps is radius check register 
db.fit_predict(data)
```

```python
# now check how it create scatterplot
sns.scatterplot(x= "data1",y="data2",data=data,hue="predict")
plt.show()
```

```python
# so it create 7 clusters 
# check we are correct or not 
sns.scatterplot(x= "data1",y="data2",data=data,hue=y)
plt.show()
```
