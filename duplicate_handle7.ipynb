```python
import pandas as pd 
data = {"Names":["a","b","c","a","b"],
        "Eng":[2,4,5,2,4],
        "Math":[4,6,9,4,6]
       }
df = pd.DataFrame(data) # ab 0 or 3 dono same ho gay or 1 and 4
df # dublicated ka hamesha pta data frame bnanan sa chlta ha 
```

```python
duplicate = df.duplicated()
df["duplicated"] = duplicate
df # ye hmna alahda sa sath column attach kr liya 
```

```python
import pandas as pd 
data = {"Names":["a","b","c","a","b"],
        "Eng":[2,4,5,2,4],
        "Math":[4,6,9,4,6]
       }
df = pd.DataFrame(data) # ab 0 or 3 dono same ho gay or 1 and 4
df # dublicated ka hamesha pta data frame bnanan sa chlta ha 
```
