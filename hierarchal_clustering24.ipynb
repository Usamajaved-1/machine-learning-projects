# Agglometric Clustering

```python
# agglometric and hierarchal_clustering only works upon linear data 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv("Iris.csv")
data.head(3)
```

```python
data.drop(columns="Species",inplace=True)
data.head(3)
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc # used to create a dendogram 
```

```python
# sc.dendrogram()
# to create a dendogram there is a need of linkage explain in register 
plt.figure(figsize=[20,20])
sc.dendrogram(sc.linkage(data,method="single",metric="euclidean")) # shift tab y is data set name method explain in 
# register  metric/matrix explain in register is used toc alculate the distance between two points 
plt.savefig("dendogram.jpg") # after save we take check the n cluster 
plt.show() #it creates a dendogram 
```

```python
# now we check th longest line in it which is at 4 ten cut into two parts explain register  
```

```python
# two clusters are created according to our dendogram 
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters=2,linkage="single") 
# two clusters are created according to our dendogram  linkage explain in register 
# agg.fit_predict(data)
data["predict"] = agg.fit_predict(data) # save our predict data into dataset 
```

```python
data.head(3)
```

```python
# now see ghow it look like now 
sns.pairplot(data = data,hue="predict")
plt.show()
```
