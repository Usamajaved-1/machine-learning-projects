```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data= pd.read_csv("placement-dataset.csv")
data.head(3)
```

```python
data.drop(columns="Unnamed: 0",inplace = True)
```

```python
data.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x = "cgpa",y = "iq",data=data,hue="placement")
plt.show()
```

```python
# we cam see data is linearlly seprable we can draw a line between them 
```

```python
x=data.iloc[:,:-1]
y=data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
p.fit(x_train,y_train)
```

```python
p.score(x_test,y_test)*100,p.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=p)
plt.show()
```
