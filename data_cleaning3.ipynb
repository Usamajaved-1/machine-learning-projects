# Missing value fill kasa krain ga 

```python
import pandas as np
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("titanic")
data.isnull().sum()
```

# data fill krna ka tariqa

```python
# data.fillna(method = "ffill",axis = 1)
data.fillna(method = "bfill",axis = 1)
data.head(20)
```

```python
mode_age = data["age"].mode()[0] 
# data["age"].fillna(mode_age,inplace = True)
data["age"] = data["age"].fillna(mode_age)
data.head()
```

```python
data.select_dtypes(include = "object")
```

```python
data.select_dtypes(include = "object").isnull().sum()
```

```python
data.select_dtypes(include = "object").columns
```

```python
for i in data.select_dtypes(include = "object").columns:
    print(i)
```

```python
# ye muyja mara sara columns ka name mil gay hain 
```

```python
for i in data.select_dtypes(include = "object").columns:
    data[i] = data[i].fillna(mode_age) #opr jo fill na ka concept lagay tha wahin copy kiya or age ko i sa change kr diya
data.select_dtypes(include = "object").isnull().sum() #data fill ho gya sara object ma check krain ga 
```
